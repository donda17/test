### Bayesian search

```python
# !pip install scikit-optimize
```

```python
def optimize(params, param_names, x, y):
    params=dict(zip(param_names, params))
    clf=ensemble.RandomForestClassifier(**params)
    kf=model_selection.StratifiedKFold(n_splits=5)
    accuracies=[]
    for ind in kf.split(X=x, y=y):
        train_ind, test_ind=ind[0], ind[1]
        xtrain=x[train_ind]
        ytrain=y[train_ind]
        
        xtest=x[test_ind]
        ytest=y[test_ind]
        
        clf.fit(xtrain, ytrain)
        preds=clf.predict(xtest)
        fold_acc=metrics.accuracy_score(ytest, preds)
        accuracies.append(fold_acc)
    
    return -1.0*np.mean(accuracies)
```

```python
from functools import partial
```

```python
from skopt import space
```

```python
param_space=[
    space.Integer(3, 15, name='max_depth'),
    space.Integer(100, 600, name='n_estimators'),
    space.Real(0.01, 1.0, prior='uniform', name='max_features'),
    space.Categorical(['gini', 'entropy'], name='criterion')
]

param_names=[
    'max_depth',
    'n_estimators',
    'max_features',
    'criterion'
]
```

```python
opt_function=partial(optimize, param_names=param_names, x=X, y=y)
```

```python
from skopt import gp_minimize
```

```python
result=gp_minimize(opt_function,
                   dimensions=param_space,
                   n_calls=15,
                   n_random_starts=5,
                   verbose=10)
```

```python
print(
    dict(zip(
        param_names,
        result.x
    ))
)
```

### [Hyperopt](https://www.kaggle.com/fanvacoolt/tutorial-on-hyperopt)

```python
from hyperopt import hp, fmin, tpe, Trials
```

```python
from hyperopt.pyll.base import scope
```

```python
def optimize(params, x, y):
    clf=ensemble.RandomForestClassifier(**params)
    kf=model_selection.StratifiedKFold(n_splits=5)
    accuracies=[]
    for ind in kf.split(X=x, y=y):
        train_ind, test_ind=ind[0], ind[1]
        xtrain=x[train_ind]
        ytrain=y[train_ind]
        
        xtest=x[test_ind]
        ytest=y[test_ind]
        
        clf.fit(xtrain, ytrain)
        preds=clf.predict(xtest)
        fold_acc=metrics.accuracy_score(ytest, preds)
        accuracies.append(fold_acc)
    
    return -1.0*np.mean(accuracies)
```

```python
param_space={
    'max_depth': scope.int(hp.quniform('max_depth', 3, 15, 1)),
    #'n_estimators': hp.quniform('n_estimators', 100, 600, 1),
    'n_estimators': scope.int(hp.quniform('n_estimators', 100, 600, 1)),
    'max_features': hp.uniform('max_features', 0.01, 1.0),
    'criterion': hp.choice('criterion', ['gini', 'entropy'])
}
```

```python
opt_function=partial(optimize, x=X, y=y)
```

```python
trials=Trials()
```

```python
result=fmin(fn=opt_function,
            space=param_space,
            algo=tpe.suggest,
            max_evals=15,
            trials=trials)
```

### optuna

```python
import optuna, optuna.trial
```

```python
def optimize(trial, x, y):
    criterion=trial.suggest_categorical('criterion', ['gini', 'entropy'])
    n_estimators=trial.suggest_int('n_estimators', 100, 600)
    max_depth=trial.suggest_int('max_depth', 3, 15)
    max_features=trial.suggest_uniform('max_features', 0.01, 1.0)

    clf=ensemble.RandomForestClassifier(
            criterion=criterion,
            n_estimators=n_estimators,
            max_depth=max_depth,
            max_features=max_features
    )
    kf=model_selection.StratifiedKFold(n_splits=5)
    accuracies=[]
    for ind in kf.split(X=x, y=y):
        train_ind, test_ind=ind[0], ind[1]
        xtrain=x[train_ind]
        ytrain=y[train_ind]
        
        xtest=x[test_ind]
        ytest=y[test_ind]
        
        clf.fit(xtrain, ytrain)
        preds=clf.predict(xtest)
        fold_acc=metrics.accuracy_score(ytest, preds)
        accuracies.append(fold_acc)
    
    return -1.0*np.mean(accuracies)
```

```python
opt_function=partial(optimize, x=X, y=y)
```

```python
study=optuna.create_study(direction='minimize')
```

```python
study.optimize(opt_function, n_trials=15)
```
